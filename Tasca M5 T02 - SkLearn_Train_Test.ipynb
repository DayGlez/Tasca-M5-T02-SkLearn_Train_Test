Ejercicio 3
Resumeix les noves columnes generades de manera estadística i gràfica

```python
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame de ejemplo con columnas dummies
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'CHAS': ['A', 'B', 'A', 'C', 'B']
})

# Seleccionar la columna categórica para crear dummies
columnas_categoricas = ['CHAS']

# Crear columnas dummies
df = pd.get_dummies(df, columns=columnas_categoricas)

# Mostrar estadísticas descriptivas
resumen_estadistico = df.describe()
print("Resumen estadístico de las nuevas columnas:")
print(resumen_estadistico)

# Mostrar gráficamente las nuevas columnas
for columna in df.columns:
    if columna.startswith('CHAS_'):  # Filtrar solo las columnas dummies generadas
        plt.figure(figsize=(6, 4))
        df[columna].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f'Distribución de valores para {columna}')
        plt.xlabel('Valor')
        plt.ylabel('Frecuencia')
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# DataFrame 
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'CHAS': ['A', 'B', 'A', 'C', 'B']
})

# Seleccionar la columna categórica para crear dummies
columnas_categoricas = ['CHAS']

# Creamos columnas dummies
df = pd.get_dummies(df, columns=columnas_categoricas)

# estadísticas descriptivas
resumen_estadistico = df.describe()
print("Resumen estadístico de las nuevas columnas:")
print(resumen_estadistico)

# Mostrar boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop('ID', axis=1))
plt.title('Boxplot para Columnas Dummies Generadas')
plt.xlabel('Columnas Dummies')
plt.ylabel('Valor')
plt.show()


# Mostrar gráficamente las nuevas columnas con scatterplot
plt.figure(figsize=(12, 6))
for columna in df.columns:
    if columna.startswith('CHAS_'):  # Filtrar solo las columnas dummies generadas
        sns.scatterplot(data=df, x='ID', y=columna, label=columna)

plt.title('Gráfico de Dispersión entre ID y Columnas Dummies Generadas')
plt.xlabel('ID')
plt.ylabel('Valor')
plt.legend()
plt.show()
```
