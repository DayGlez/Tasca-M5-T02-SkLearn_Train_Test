### Estandarización de datos númericos

En este apartado vamos a estandarizar las columnas númericas, CHAS y RAD no las incluiremos ya que son variables categóricas. 

Previamente a realizar la estandarización aplicaremos Test de normalidad a las variables, para ver si se distribuyen como una normal o no.  

### Test Shapiro

Planteamos la hipotesis
- hipotesis nula H0: tiende a una distribución normal
- hipotesis alternativa H1: no tiende a una distribución normal


- Si Pvalor >= 0.05(nivel de signficación)-->No Rho
- Si Pvalor < 0.05 (nivel de significacion)-->Rho

```python
from scipy.stats import shapiro

# Seleccionamos el conjunto de datos X_Train
variables_numericas_train = X_train

# Inicializamos las listas
resultados_shapiro = []
rechaza_hipotesis_nula_por_variable = []

# Realizamos el test de normalidad de Shapiro-Wilk para X_train
for columna in variables_numericas_train.columns:
    stat, p_value_shapiro = shapiro(variables_numericas_train[columna])
    resultados_shapiro.append((columna, p_value_shapiro))

    # Verificamos si rechazamos la hipótesis nula con un nivel de significación del 5%
    alfa = 0.05
    rechaza_hipotesis_nula = p_value_shapiro < alfa
    rechaza_hipotesis_nula_por_variable.append((columna, rechaza_hipotesis_nula))

# Imprimimos los resultados
for resultado, rechazo in zip(resultados_shapiro, rechaza_hipotesis_nula_por_variable):
    print(f'Variable: {resultado[0]}, p-valor: {resultado[1]}, ¿Rechazamos la H0? {rechazo[1]}')
```

Las variables de estudio rechazan la H0, por lo tanto no se distribuye como una normal y no procederemos a estandarizarlo.

```python
from scipy.stats import shapiro

# Seleccionamos un atributo del conjunto de datos X_Test
variables_numericas_train = X_test

# Realizamos el test de normalidad de Shapiro-Wilk para X_Test
for columna in variables_numericas_train.columns:
    stat, p_value_shapiro = shapiro(variables_numericas_train[columna])
    resultados_shapiro.append((columna, p_value_shapiro))
    
# Verificamos si rechazamos la hipótesis nula con un nivel de significación del 5%
    alfa = 0.05
    rechaza_hipotesis_nula = p_value_shapiro < alfa
    rechaza_hipotesis_nula_por_variable.append((columna, rechaza_hipotesis_nula))

# Imprimimos los resultados
for resultado, rechazo in zip(resultados_shapiro, rechaza_hipotesis_nula_por_variable):
    print(f'Variable: {resultado[0]}, p-valor: {resultado[1]}, ¿Rechazamos la H0? {rechazo[1]}')
```

Las variables de estudio rechazan la H0, por lo tanto no se distribuye como una normal y no procederemos a estandarizarlo.

Clasificamos las columnas entre outliers y no outliers. 

Analizamos X_train

```python
# Creamos un diccionario para para almacenar las columnas clasificadas como outliers o no outliers
outliers_diccionario = {} 

for columna in X_train.columns:
    # Asignamos los valores a los cuartiles
    Q1 = X_train[columna].quantile(0.25)
    Q3 = X_train[columna].quantile(0.75)

    # rango intercuartílico (IQR)
    IQR = Q3 - Q1

    # Calculamos los límites para los bigotes
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Identificar los outliers
    outliers = X_train[(X_train[columna] < lower_limit) | (X_train[columna] > upper_limit)]

    # Clasificar la columna como outliers o no outliers
    if outliers.empty:
        outliers_diccionario[columna] = 'No Outliers'
    else:
        outliers_diccionario[columna] = 'Outliers'

# Imprimir la clasificación
for columna, clasificacion in outliers_diccionario.items():
    print(f'{columna}: {clasificacion}')

```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler

# Aplicar escala min-max a los valores sin outliers
X_no_outliers_minmax = X_train.copy()
for columna in X_train.columns:
    if outliers_diccionario[columna] == 'No Outliers':
        min_max_scaler = MinMaxScaler()
        X_no_outliers_minmax[columna] = min_max_scaler.fit_transform(X_train[[columna]])

# Graficar valores sin outliers con escala min-max
X_no_outliers_minmax.boxplot(figsize=(12, 8))
plt.title('Boxplot sin outliers con escala min-max')
plt.show()

# Aplicar escala robusta a los valores con outliers
robust_scaler = RobustScaler()
X_with_outliers_robust = pd.DataFrame(robust_scaler.fit_transform(X_train), columns=X_train.columns)

# Graficar valores con outliers con escala robusta
X_with_outliers_robust.boxplot(figsize=(12, 8))
plt.title('Boxplot con outliers con escala robusta')
plt.show()
```

El uso de Robust y minmax lo utilizamos con el objetivo de visualizar como cambian las distribuciones de las variables después de aplicar diferentes técnicas de escala en el conjunto de X_train y X_test con y sin outliers.

En el primer caso, Robust, el prposito principal es escalar los datos de manera robusta frente a los outliers, es decir, que el conjunto de datos no se vea afectado de manera significativa por valores atipicos. Por otra parte, minmax, es útil cuando se desea que todas las variables estén en el mismo rango y se encuentren en una escala especifca. Aquí deseamos comparar variables en la misma escala cuando la presencia de valores atipicos/outliers no es un problema. 

En el boxplot anterior podemos oberservar valores atipicos/outliers cercanos y dispersos. Este nos indidca  que los valores o magnitudes sos más similares entre si, cercanos, y dispersos todo lo contrario, que tienen magnitud muy diferente entre sí.
